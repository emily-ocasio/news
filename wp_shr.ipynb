```python
import pandas as pd
```

```python
shr_file = "SHR65_23.csv"
shr = pd.read_csv(shr_file)
shr
```

```python
len(shr)
```

```python
shr_dc = shr[shr['State'] == 'District of Columbia']
shr_dc
```

```python
shr_dc.columns
```

```python
shr_dc['Year'].value_counts()
```

```python
shr_dc['Year'].unique()
```

```python
shr_dc = shr_dc[shr_dc['Year'] >= 1977]
shr_dc = shr_dc[shr_dc['Year'] <= 2010]
shr_dc
```

```python
shr_dc['VicRace'].value_counts()
```

```python
dates = pd.to_datetime([f"{m}-1-{y}" for y,m in zip(shr_dc.Year, shr_dc.Month)])
dates
```

```python
date_periods = [pd.Period(freq='m', year = y, month = m) for m,y in zip(dates.month, dates.year)]
date_periods
```

```python
shr_dc['YearMonth'] = date_periods
shr_dc['YearMonth']
```

```python
shr_dc[['Year', 'Month', 'YearMonth']]
```

```python
shr_dc.YearMonth = shr_dc.YearMonth.astype(str)
```

```python
import sqlite3
con = sqlite3.connect('newarticles.db')
```

```python
shr_dc.to_sql('shr', if_exists='append', con=con)
```
